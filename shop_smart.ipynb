**Feature Engineering Strategy: VisitorType**

We will encode the `VisitorType` column using an ordinal mapping to capture user engagement levels.

* **0 (Other):** Lowest priority or unclassified traffic.
* **1 (New_Visitor):** Standard potential, first-time engagement.
* **2 (Returning_Visitor):** Highest priority, representing loyal or repeat users.

**Reasoning:**
This ordinal approach allows the Decision Tree to utilize the inherent hierarchy of user value (e.g., splitting on `VisitorType >= 2` to isolate high-value returning customers) without creating multiple sparse columns via One-Hot Encoding.

```python
df['Admin_Avg_Duration'] = np.where(
    df['Administrative'] == 0, 
    0, 
    df['Administrative_Duration'] / df['Administrative']
)
```

```python
df['Info_Avg_Duration'] = np.where(
    df['Informational'] == 0, 
    0, 
    df['Informational_Duration'] / df['Informational']
)
```

```python
df['Product_Avg_Duration'] = np.where(
    df['ProductRelated'] == 0, 
    0, 
    df['ProductRelated_Duration'] / df['ProductRelated']
)
```

```python
# Drop the "Total Duration" columns because they are now redundant
# We keep the "Counts" (Volume) and the "New Averages" (Intensity)
cols_to_drop = ['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration']
df.drop(columns=cols_to_drop, inplace=True)

# Check the new columns
print(df.columns)
```

```python
# Define the 3 new features
new_features = ['Admin_Avg_Duration', 'Info_Avg_Duration', 'Product_Avg_Duration']

# Create subplots: 1 Row, 3 Columns
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Loop through the features and plot them on their respective axes
# Updated loop to fix the FutureWarning
for i, col in enumerate(new_features):
    # Fix: Added hue='Revenue' and legend=False
    sns.barplot(data=df, x='Revenue', y=col, hue='Revenue', ax=axes[i], palette='viridis', legend=False)
    
    axes[i].set_title(f'Avg Duration: {col} vs Revenue')
    axes[i].set_ylabel('Average Duration (seconds)')
    axes[i].set_xlabel('Revenue (Purchase)')

plt.tight_layout()
plt.show()
```

**EDA Conclusion: Average Duration Features**

**Observation:**
We compared the Average Duration per page type against the target variable `Revenue`.
1.  **Consistent Pattern:** In all three categories (Admin, Info, Product), purchasing users spend significantly more time per page on average.
2.  **Strongest Signal:** The `Informational` duration shows the largest relative increase (~70% higher for buyers), while `Product` duration shows the highest absolute time spent.

**Decision:**
These engineered features (`*_Avg_Duration`) are strong discriminators of user intent. We will retain them for the Decision Tree model, as they capture engagement intensity better than raw page counts alone.

# **EDA Conclusion & Feature Strategy**

After analyzing the relationships between various features and the target variable (`Revenue`), we have identified the key drivers of user purchasing behavior and filtered out non-predictive "noise."

### **1. Key Drivers of Purchase (Strong Signals)**
* **PageValues (Critical):** This is the strongest predictor in the dataset. Users who visit high-value pages (e.g., checkout, cart) show a massive increase in conversion probability. This will likely be the root node of our Decision Tree.
* **Bounce & Exit Rates:** Purchasing users exhibit significantly lower bounce and exit rates. This confirms that "sticky" users who browse multiple pages are more likely to convert.
* **Seasonality (Month):** We observed distinct spikes in purchase activity during **November (Festive)**, **May (Summer/Travel)**, and **December**, making `Month` a crucial categorical feature.

### **2. Feature Engineering: Engagement Intensity**
Instead of using raw "Total Duration" (which is redundant with Page Counts), we engineered **Average Duration** features (`Admin_Avg_Duration`, `Info_Avg_Duration`, `Product_Avg_Duration`).
* **Insight:** Buyers consistently spend more time *per page* than non-buyers.
* **Result:** `Product_Avg_Duration` showed the clearest separation, indicating that users who carefully read product details are high-intent buyers.

### **3. Dropped Features (Noise Reduction)**
We will remove the following features because they showed **near-zero correlation** with Revenue (flat trend lines):
* `Region`, `OperatingSystems`, `Browser`: Technical/Demographic attributes did not meaningfully impact conversion rates.
* `SpecialDay`: This feature failed to capture the actual seasonal trends (unlike `Month`) and was deemed ineffective.



# Feature Encoding

```python
# 1. Define the months we saw in our EDA as "Peak"
peak_months = ['Nov', 'Dec', 'May', 'Oct']

# 2. Apply the logic: 
# "Return 1 if month is in the list, else 0"
df['Is_Peak_Season'] = df['Month'].apply(lambda x: 1 if x in peak_months else 0)

# 3. Drop the old Month column (we don't need it now)
df.drop(columns=['Month'], inplace=True)
```

```python
# 1. Define the mapping dictionary
# This tells pandas exactly what number to swap for each text label
visitor_mapping = {
    'Other': 0,
    'New_Visitor': 1,
    'Returning_Visitor': 2
}

# 2. Apply the mapping
df['VisitorType'] = df['VisitorType'].map(visitor_mapping)
```

```python
# Convert boolean columns to 0 and 1
df['Weekend'] = df['Weekend'].astype(int)
df['Revenue'] = df['Revenue'].astype(int)
```

# Train-Test-Split

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns = ['SpecialDay', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'Revenue'])
y = df['Revenue']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

# Model Training

### Decision Tree

```python
# Decision tree - no pruning
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score
y_pred = model.predict(X_test)

print(f'accuraacy: {accuracy_score(y_test, y_pred)}')
print(f'precision: {precision_score(y_test, y_pred)}')
print(f'f1: {f1_score(y_test, y_pred)}')
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize = (18,10))
plot_tree(
    model,
    feature_names = X.columns,
    class_names = ['Died', 'Survived'],
    filled = True,
)

plt.tight_layout()
plt.show()
```

# Decision Tree with pre-pruning

```python
max_depth = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for depth in max_depth:
    model = DecisionTreeClassifier(max_depth=depth)
    model.fit(X_train, y_train)

    acc = model.score(X_test, y_test)
    print(f"For depth = {depth}, accuracy = {acc}")
```

```python
# max_depth = 4 is the best
model = DecisionTreeClassifier(max_depth = 4)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(f'accuraacy: {accuracy_score(y_test, y_pred)}')
print(f'precision: {precision_score(y_test, y_pred)}')
print(f'f1: {f1_score(y_test, y_pred)}')
```

```python
min_sample_splits = [5, 10, 15, 25, 30]

for splits in min_sample_splits:
    model = DecisionTreeClassifier(max_depth = 4, min_samples_split = splits)
    model.fit(X_train, y_train)

    acc = model.score(X_test, y_test)
    print(f"For sample splits= {splits}, accuracy = {acc} ")
```

# Decision tree with post-pruning

```python
full_tree = DecisionTreeClassifier(random_state = 42)
full_tree.fit(X_train, y_train)
```

```python
path = full_tree.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

print(ccp_alphas)
```

```python
# Train our model for all alphas

tree = []

for alpha in ccp_alphas:
    model = DecisionTreeClassifier(random_state = 42, ccp_alpha = alpha)
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)

    tree.append((model, alpha))
```

```python
best_acc = 0
best_alpha = 0

for model, alpha in tree:
    acc = model.score(X_test, y_test)
    if acc > best_acc:
        best_acc = acc
        best_alpha = alpha
```

```python
best_acc
```

```python
best_alpha
```

```python
best_model = DecisionTreeClassifier(ccp_alpha = best_alpha, max_depth = 4)
best_model.fit(X_train, y_train)
```

```python
y_pred = best_model.predict(X_test)

print(f'accuraacy: {accuracy_score(y_test, y_pred)}')
print(f'precision: {precision_score(y_test, y_pred)}')
print(f'f1: {f1_score(y_test, y_pred)}')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [4, 6, 8, 10, None],
    'min_samples_split': [2, 10, 20, 50],
    'criterion': ['gini', 'entropy']
}

grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, scoring='f1')
grid_search.fit(X_train, y_train)
print(f"Best Params: {grid_search.best_params_}")
```
